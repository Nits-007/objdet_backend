```python
import cv2
import time

# Initialize video capture
cap = cv2.VideoCapture('vid.mp4')

# Check if video capture is successful
if not cap.isOpened():
    print('Unable to open video')
    exit()

# Initialize variables for FPS calculation
fps = cap.get(cv2.CAP_PROP_FPS)
fps_display = 0  # Initialize fps_display
fps_display_interval = 1  # Interval in seconds to update FPS display
last_time = time.time()
frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        print('Unable to read video')
        break

    # Get current time
    current_time = time.time()

    # Calculate elapsed time
    elapsed_time = current_time - last_time
    frame_count += 1

    # Update FPS display if the interval has passed
    if elapsed_time > fps_display_interval:
        fps_display = frame_count / elapsed_time
        last_time = current_time
        frame_count = 0

    # Get YOLO predictions
    pred_image = yolo.predictions(frame)

    # Display FPS on the frame
    cv2.putText(pred_image, f'FPS: {fps_display:.2f}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    # Show the frame with predictions
    cv2.imshow('YOLO', pred_image)
    
    # Exit loop on pressing 'Esc' key
    if cv2.waitKey(1) == 27:
        break

# Release video capture and destroy all windows
cap.release()
cv2.destroyAllWindows()
```
